2. Pongamos en práctica el loc e iloc:

Usando el loc:

Extraed todas las filas de la columna "price". ¿De qué tipo es el resultado obtenido?

```python
price = fichero.loc[:,'price']
price.head(5)

# Nos devuelve una serie
```

Extraed todas las filas de las columnas "price" y "retail_price". ¿De qué tipo es el resultado obtenido?

```python
price_retail = fichero.loc[:,['price','retail_price']]
price_retail.head(5)
```

Extraed las filas pares, de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count".

```python
list(fichero.columns).index("rating") # para saber el índice hay que convertir en lista para aplicar index
```

```python
pares = fichero.loc[0::2,['rating','rating_count', 'rating_five_count', 'rating_four_count', 'rating_three_count', 'rating_two_count', 'rating_one_count']]
pares.head(6)
```

Extraed las las filas cuyo precio sea mayor que 10 y sus unidades vendidas ("units_sold") sea mayor que 5000.

```python
ventas = fichero.loc[(fichero.price>10) & (fichero.units_sold>5000)]
ventas.head(5)
```
